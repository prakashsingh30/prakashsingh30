```python
s=[]
for i in df['Sequence'].values:
    s.append(i.split(' ,')[1])
  
# Storing Sequence in s[0] ,s[1] .
```

```python
y, n= 0,0
lst=[1,2,3,4,5,6,7,8]
for a in lst:
    for j in s:
      if str(a) in j:
        y=y+1
      else:
        n=n+1
    print(a ,"Freqency :" ,(y/(y+n))*100)
```

```python
x='M-M'
df_new=df[df['Category']==x]
data1 = df_new['Sequence'].values

lst=[1,2,3,4,5,6,7,8]


for j in lst:
    y,n=0,0
    for i in data1:
       t=i.split(' ,')
       if str(j) in t[0]:   # 1st index of sequence .
         y=y+1
       else:
         n=n+1
    print(j ," :", (y/(y+n))*100)
```
